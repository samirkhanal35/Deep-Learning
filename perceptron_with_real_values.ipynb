```python
import random
```

```python
no_of_data = 1000
input_data = []
output_data = []
for each in range(no_of_data):
    w1 = round(random.uniform(0.0, 1.0), 1)
    w2 = round(random.uniform(0.0, 1.0), 1)
    input_data.append((w1, w2))
    if w1>=0.5 and w2>=0.5:
        output_data.append(1)
    else:
        output_data.append(0)

input_data
```

```python
output_data
```

```python
# learning rate
lr = 0.1

# Number of trial => epochs
epochs = 1000
```

```python
# Initialize weights and bias
w1 = random.uniform(-1 , 1)
w2 = random.uniform(-1 , 1)
b = random.uniform(-1 , 1)
w1, w2, b
```

```python
# Define activation function => step function
def activate(sum):
    threshold = 0
    if sum>=0: 
        return 1
    else:
        return 0
```

```python
for epoch in range(epochs):
    total_error = 0

    for i in range(len(input_data)):
        input1, input2 = input_data[i]
        target_output = output_data[i]

        # Compute weighted sum
        weighted_sum = input1 * w1 + input2 * w2 + b

        output = activate(weighted_sum)
        print("output:", output)
                
        error = target_output - output 
        print("error:", error)


        # update weights and bias
        w1 += lr * error * input1 
        w2 += lr * error * input2 
        b += lr * error

        total_error = abs(error) 
        print("Total error:", error)

    if total_error == 0:
        print("Epoch number:", epoch)
        print("Final weights:", w1, w2)
        break
```

```python
# Testing 
test_data = [(0,0), (0,1), (1,0),(1,1), (0.5, 0), (0, 0.5), (0.2, 0.2), (0.5, 0.5), (1, 0.5), (0.7, 0.8), (0.3, 0.7)]

for e_inp_data in test_data:
    input1, input2 = e_inp_data

    # Compute weighted sum
    weighted_sum = input1 * w1 + input2 * w2 + b

    output = activate(weighted_sum)
    print("input1:",input1,"input2:", input2,"output:", output)
```
